```python
import csv

# 해당 코드의 로직을 slices_labeling.py로
# 원본 CSV 파일 경로 (현재 디렉토리에 있는 SARD_SINK.csv 파일)
input_file = "SARD_Sink_label_info_version_Buffer.csv"

# CSV 파일 읽기
with open(input_file, "r", encoding="utf-8") as infile:
    lines = infile.readlines()

# Dictionary 형태로 저장할 데이터
data_dict = {}
current_file = None
current_sink = []

for line in lines:
    line = line.rstrip()  # 오른쪽 개행 제거
    
    # 새로운 File# (CWE...)이 등장하면 저장 후 초기화
    if line.startswith("CWE"):
        if current_file is not None:
            data_dict[current_file] = "\n".join(current_sink)  # 개행 유지하여 저장
        
        parts = line.split(",", 1)  # 첫 번째 쉼표를 기준으로 File#과 sink_keyword 분리
        current_file = parts[0]
        current_sink = [parts[1]] if len(parts) > 1 else []
    
    else:
        # 기존 sink_keyword에 이어붙이기 (개행 유지)
        if current_sink:
            current_sink.append(line)

# 마지막 데이터 추가
if current_file is not None:
    data_dict[current_file] = "\n".join(current_sink)  # 개행 유지하여 저장

# 변환된 데이터 확인
for key, value in data_dict.items():
    print(f"{key}:\n{value}\n{'-'*40}")
```
