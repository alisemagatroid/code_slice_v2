```python
import csv

#    slices_labeling.py
#  CSV   (   SARD_SINK.csv )
input_file = "SARD_Sink_label_info_version_Buffer.csv"

# CSV  
with open(input_file, "r", encoding="utf-8") as infile:
    lines = infile.readlines()

# Dictionary   
data_dict = {}
current_file = None
current_sink = []

for line in lines:
    line = line.rstrip()  #   
    
    #  File# (CWE...)    
    if line.startswith("CWE"):
        if current_file is not None:
            data_dict[current_file] = "\n".join(current_sink)  #   
        
        parts = line.split(",", 1)  #     File# sink_keyword 
        current_file = parts[0]
        current_sink = [parts[1]] if len(parts) > 1 else []
    
    else:
        #  sink_keyword  ( )
        if current_sink:
            current_sink.append(line)

#   
if current_file is not None:
    data_dict[current_file] = "\n".join(current_sink)  #   

#   
for key, value in data_dict.items():
    print(f"{key}:\n{value}\n{'-'*40}")
```
